```python
import pandas as pd
```

```python
df = pd.read_csv("data/genotype-phenotype-assertions-ALL.csv")
df.head()
```

```python
gmt = {}
genes = set()
for i in df.iterrows():
    data = i[1]
    mp_term_name = data['mp_term_name']
    mp_term_id = data['mp_term_id']
    label = "%s (%s)\t"%(mp_term_name, mp_term_id)
    if label not in gmt:
        gmt[label] = set()
    
    gene = data['marker_symbol']
    gene = gene.upper()
    genes.add(gene)
    gmt[label].add(gene)

print("Terms: %s"%len(gmt))

print("Gene Coverage: %s"%len(genes))
gene_counts = 0
max_genes = 0
min_genes = 100
for i in gmt.values():
    gene_counts += len(i)
    if len(i) > max_genes:
        max_genes = len(i)
    if len(i) < min_genes:
        min_genes = len(i)

print("Average genes per terms: %s"%(gene_counts/len(gmt)))
print("Max genes per terms: %s"%(max_genes))
print("Min genes per terms: %s"%(min_genes))

```

```python
with open("/Users/maayanlab/Documents/codebase/enrichr_utils/gmt/KOMP2.gmt", "w") as o:
    for k,v in gmt.items():
        line = [k] + list(v) + ["\n"]
        o.write("\t".join(line))
    
```
