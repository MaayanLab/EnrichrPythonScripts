## build dictionary of dieases and associated genes

```python
phenotypes = {}

for i in range(0, len(phenotype)):
    p = phenotype.iloc[i, 4]
    if p in phenotypes.keys():
        phenotypes[p].append(phenotype.iloc[i,2])
    else:
        phenotypes[p] = [phenotype.iloc[i,2]]
```

```python
len(phenotypes['WBPhenotype:0000643'])
```

## get the ontologies for phenotypes

```python
graph = obonet.read_obo('in_production/phenotype_ontology.WS266.obo')
term_data = graph.nodes(data=True)
term_data['WBPhenotype:0000000']['name']
```

## build gmt

```python
filename = 'phenotype_association_wormbase_%s.gmt'% str(datetime.date.today())[0:7].replace('-', '_')
file = open(filename,'w+') 
terms = phenotypes.keys()
for i,term in enumerate(terms):
    
    progressPercent = ((i+1)/len(terms))*100
    sys.stdout.write("Progress: %d%%  %d Out of %d   \r" % (progressPercent, (i+1), len(terms)))
    sys.stdout.flush()
    
    if len(set(phenotypes[term])) > 4:
        name = term_data[term]['name'] + '_' + term
        file.write("%s\t" % name)
        file.write("\t")
        genes = set(phenotypes[term])
    
        for gene in genes:
              file.write("%s\t" % gene)
        file.write("\n")
        
file.close()
```
